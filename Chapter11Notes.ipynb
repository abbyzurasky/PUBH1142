### Import Packages

```python
# import the numpy package
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

### Creating a Population

```python
# Seed the numpy psuedo-random number generator
np.random.seed(28)

# Create an array of 1000 values using a continuous uniform distribution between 50 and 100 and assign it to the variable population
population = np.random.uniform(50, 100, 10000)
```

### Create a graphical representation of the population

```python
# Create a histogram with seaborn of the data in population
sns.histplot(population, kde=False)
```

### Calculate some statistics for the population

```python
# Calculate the mean of the population and assign it to the variable population_mean
population_mean = population.mean()
print(population_mean)

# Calculate the standard deviation of the population and assign it to the variable population_std
population_std = population.std()
population_std
```

### Calculate some statistics of a sample of the population

```python
# Select a random sample of 30 values from the population and assign it to the variable sample
sample = np.random.choice(population, 30)

# Calculate the mean of the sample and assign it to the variable sample_mean
sample_mean = sample.mean()
sample_mean
```

### Calculate the sample mean of 1000 samples of the size 30

```python
# Use list comprehension to calculate the sample mean of 1000 samples of size 30 and assign it to the variable sample_means
sample_means = [np.random.choice(population, 30).mean() for i in range(1000)]
sample_means
```

### Create a graphical representation of the data above

```python
# Create a histogram with seaborn of the data in sample_means
sns.histplot(sample_means, kde=False)
```

```python
# Sort the values in sample_means and assign them to the variable sorted_sample_means
sorted_sample_means = np.sort(sample_means)

# Calculate the value at the 25th position
print(sorted_sample_means[24])

# Calculate the value at the 975th position
sorted_sample_means[974]
```
